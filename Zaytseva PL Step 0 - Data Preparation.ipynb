```python
import pandas as pd
```

```python
links = ['https://www.youtube.com/watch?v=kd_TNFN43fo&t=461s', 'https://www.youtube.com/watch?v=aKLEh3UkR4o', 'https://www.youtube.com/watch?v=B08UvVwjpG4', 'https://www.youtube.com/watch?v=H_-iU6AW1c0', 'https://www.youtube.com/watch?v=TMmDebW_OBI','https://www.youtube.com/watch?v=nv3iOH_XYdo']
```

```python
# Extract ids from links

def get_id (url_ids):
    ids = []
    for link in links:
        id = link[32:]
        ids.append(id)
    return ids
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
url_ids = get_id(links)
url_ids
```

```python
len(url_ids)
```

```python
# Extract subtitles
script = []
def get_data (ids):
    for id in url_ids:
        data = YouTubeTranscriptApi.get_transcript(id)
        script.append(data)
    return script
```

```python
get_data (url_ids)
```

Вынимаем из субтитров текст и убираем "вспомогательные" пояснения ([Music])

```python
# Get texts

texts = []
for i in script:
    list_text = []
    for j in i:
        list_text.append(j['text'])
        a = " ".join(list_text)
    texts.append(a)
texts
```

```python
import re
clear_texts = []
for text in texts:
    clear = re.sub(r'\[Music\]', '', text).strip()
    clear_texts.append(clear)
clear_texts
```

```python
# Dividing texts by sentences

model, example_texts, languages, punct, apply_te = torch.hub.load(repo_or_dir='snakers4/silero-models',
                                                                  model='silero_te')
new_texts = []
for text in clear_texts:
    input_text = text
    text_punct = apply_te(input_text, lan='en')
    new_texts.append(text_punct)
new_texts
```

Поместим тексты в датафрейм и создадим файл 

```python
df = pd.DataFrame(new_texts, columns = ['texts'])
df
```
